```python
!pip install pyspark
import pyspark
from pyspark import SparkContext
from pyspark.sql import SparkSession

spark = SparkSession.builder.master("local").appName("Store market Analysis").getOrCreate()

StocksAMRB = spark.read.option("header",True) \
     .csv("AMRB.csv")

StocksAMRB.show()
```

```python
StocksAMRB.createOrReplaceTempView("AMRB");
```

```python
spark.sql("select count(*) from AMRB").show()
```

```python
spark.sql("select * from AMRB")
```

```python
spark.sql("select Date,Open, High,Low,Close,'Adj Close',Volume from AMRB").show()
```

```python
# Transactions Quotidiennes (facile)
spark.sql("select Date,Open,Close from AMRB").show()
```

```python
# Volatilité Journalière (facile)
spark.sql("select Date, (Close - Open) As VolatiliteJournaliere from AMRB").show()
```

```python
# Extrêmes de Volume (intermédiaire)
spark.sql("select min(Volume) As MinVolume, max(Volume) As MaxVolume from AMRB").show()
```

```python
# Moyenne Annuelle des Valeurs d'Ouverture (intermédiaire)
spark.sql("select YEAR(Date) As Annee, avg(Open) As Moyenne from AMRB Group By YEAR(Date)").show()
```

```python
# Total des Volumes par Mois (intermédiaire)
spark.sql("select MONTH(Date) As Mois, sum(Volume) As TotalVolume from AMRB Group by MONTH(Date) Order by MONTH(Date)").show()
```

```python
spark.sql("select (Close - Open) As Ecart from AMRB Order By (Close - Open) DESC").show()
```

```python
# Plus grand ecart quotidien
spark.sql("select Ecart As GrandEcart from (select (Close - Open) As Ecart from AMRB Order By (Close - Open) Desc ) Limit 1").show()
```

```python
# Moyenne Mobile du prix de Cloture sur 7 Jours (avancé)
from pyspark.sql import Window
from pyspark.sql import functions as F


w = (Window.orderBy('Date').rowsBetween(-6, 0))

StocksAMRB_new = StocksAMRB.withColumn('rolling_average', F.avg('Close').over(w))

StocksAMRB_new.show()
```
